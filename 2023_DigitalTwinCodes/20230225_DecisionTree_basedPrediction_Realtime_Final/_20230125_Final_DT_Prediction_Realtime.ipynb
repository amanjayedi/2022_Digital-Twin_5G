```python
import numpy as np
import pandas as pd
import joblib

# Import ML-based prediction model
prediction_model = joblib.load("/content/drive/MyDrive/MIZZOU/DigitalTwin_AT&T/_2023_DigitalTwinCodes/20230225_DecisionTree_basedPrediction_Realtime_Final/DigitalTwin_DT.joblib")

# Load data from csv and get the last row for prediction
data = pd.read_csv("/content/drive/MyDrive/MIZZOU/DigitalTwin_AT&T/_2023_DigitalTwinCodes/20230225_DecisionTree_basedPrediction_Realtime_Final/combined_weather_sensor.csv", usecols=['Temperature (degC)', 'Humidity (%)', 'Wind Speed (m/s)', 'Wind Direction (deg)', 'Pressure (hPa)', 'PeoplePerArea(p/m2)', 'LightingPowerIntensity(W/m2)', 'EquipmentPowerIntensity(W/m2)', 'ACH'])
data_for_prediction = data.tail(1)

# Perform real-time prediction using the ML-based model
prediction_realtime = prediction_model.predict(data_for_prediction)

# Append the predicted value to the Carbon Emissions (lb-co2/h) column of the same row
data.loc[data.index[-1], 'Carbon Emissions (lb-co2/h)'] = prediction_realtime[0]

# Save the updated data to the same csv file
data.to_csv("/content/drive/MyDrive/MIZZOU/DigitalTwin_AT&T/_2023_DigitalTwinCodes/20230225_DecisionTree_basedPrediction_Realtime_Final/combined_weather_sensor.csv", index=False)
print("data is appended to combined csv file after ML based prediction")
```

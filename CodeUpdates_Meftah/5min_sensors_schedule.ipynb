```python
dates = pd.date_range('2023-01-01', '2023-12-31', freq='D')
drange = range(len(dates))
#Converting to series for getting weekdays name
s = dates.to_series(index=drange)
# dayofyear = s.dt.datetime
y_day = list(dates)

weekdays = s.dt.day_name()
w_day = list(weekdays)
# print(w_day)
```

```python
day_tuple = list(zip(y_day, w_day))

people = []
min = 60 # 1 min interval of an hour data
hours_of_day = 24*min  

for date, weekday in day_tuple:

  #Before spring semester
  if date >= datetime(2023,1,1) and date < datetime(2023,1,17):
    if weekday == 'Saturday' or weekday == 'Sunday':
      for hr in range(hours_of_day):
        people.append(0)
    else:
      for hr in range(hours_of_day):
         if hr>=0 and hr<10*min:
            people.append(0)
         elif hr>=10*min and hr<17*min:
            people.append(random.randint(0,2))
         else:
            people.append(0)

  #Before spring break
  elif date >= datetime(2023,1,17) and date < datetime(2023,3,26):
    if weekday == 'Saturday' or weekday == 'Sunday':
      for hr in range(hours_of_day):
        if hr<7*min:
          people.append(0)
        elif hr>=7*min and hr<16*min:
          people.append(random.randint(0,3))
        else:
          people.append(random.randint(0,9))    
    elif weekday == 'Monday' or weekday == 'Wednesday' or weekday == 'Friday':
      for hr in range(hours_of_day):
        if hr>=0 and hr<7*min:
            people.append(0)
        elif hr>=7*min and hr<9*min:
            people.append(random.randint(0,10))
        elif hr>=9*min and hr<=13*min:
            people.append(random.randint(30,32))
        elif hr>13*min and hr<=17*min:
          people.append(random.randint(14,16))
        else:
          people.append(random.randint(0,3))
    elif weekday == 'Tuesday':
      for hr in range(hours_of_day):
         if hr>=0 and hr<7*min:
            people.append(0)
         elif hr>=7*min and hr<9*min:
            people.append(random.randint(0,2))
         elif hr>=9*min and hr<=11*min:
            people.append(5)
         elif hr>11 and hr<=17*min:
           people.append(random.randint(23,25))
         else:
           people.append(random.randint(0,3))
    elif weekday == 'Thursday':
      for hr in range(hours_of_day):
         if hr>=0 and hr<7*min:
            people.append(0)
         elif hr>=7*min and hr<9*min:
            people.append(random.randint(0,5))
         elif hr>=9*min and hr<=11*min:
            people.append(random.randint(52,55))
         elif hr>13*min and hr<=17*min:
           people.append(random.randint(23,25))
         else:
           people.append(random.randint(0,5))
 
  #Spring break
  elif date >= datetime(2023,3,26) and date < datetime(2023,4,4):
    for hr in range(hours_of_day):
      people.append(0)  

  #After spring break to exam class end date 
  elif date >= datetime(2023,4,4) and date< datetime(2023,5,6):
      if weekday == 'Saturday' or weekday == 'Sunday':
        for hr in range(hours_of_day):
          if hr<7*min:
            people.append(0)
          elif hr>=7*min and hr<16*min:
            people.append(random.randint(0,3))
          else:
            people.append(random.randint(0,9))    
      elif weekday == 'Monday' or weekday == 'Wednesday' or weekday == 'Friday':
        for hr in range(hours_of_day):
          if hr>=0 and hr<7*min:
              people.append(0)
          elif hr>=7*min and hr<9*min:
              people.append(random.randint(0,10))
          elif hr>=9*min and hr<=13*min:
              people.append(random.randint(30,32))
          elif hr>13*min and hr<=17*min:
            people.append(random.randint(14,16))
          else:
            people.append(random.randint(0,3))
      elif weekday == 'Tuesday':
        for hr in range(hours_of_day):
          if hr>=0 and hr<6*min:
              people.append(0)
          elif hr>=6*min and hr<9*min:
              people.append(random.randint(0,2))
          elif hr>=9*min and hr<=11*min:
              people.append(5)
          elif hr>11*min and hr<=17*min:
            people.append(random.randint(23,25))
          else:
            people.append(random.randint(0,5))
      else:
        for hr in range(hours_of_day):
          if hr>=0 and hr<6*min:
              people.append(0)
          elif hr>=6*min and hr<9*min:
              people.append(random.randint(0,5))
          elif hr>=9*min and hr<=11*min:
              people.append(random.randint(50,55))
          elif hr>13*min and hr<=17*min:
            people.append(random.randint(23,25))
          else:
            people.append(0)
 
  #Spring semseter exam weeks
  elif date >= datetime(2023,5,6) and date< datetime(2023,5,17):
    if weekday == 'Saturday' or weekday == 'Sunday':
      for hr in range(hours_of_day):
        people.append(0)
    else:
      for hr in range(hours_of_day):
         if hr>=0 and hr<10*min:
            people.append(0)
         elif hr>=10*min and hr<17*min:
            people.append(random.randint(0,5))
         else:
            people.append(0)
  
  #Summer break
  elif date >= datetime(2023,5,17) and date < datetime(2023,8,17):
    for hr in range(hours_of_day):
      people.append(0)  
 
  #Fall semseter before thanksgiving break
  elif date >= datetime(2023,8,17) and date < datetime(2023,11,21):
    if weekday == 'Saturday' or weekday == 'Sunday':
      for hr in range(hours_of_day):
        if hr<7*min:
          people.append(0)
        elif hr>=7*min and hr<16*min:
          people.append(random.randint(0,3))
        else:
          people.append(random.randint(0,9))    
    elif weekday == 'Monday' or weekday == 'Wednesday' or weekday == 'Friday':
      for hr in range(hours_of_day):
        if hr>=0 and hr<7*min:
            people.append(0)
        elif hr>=7*min and hr<9*min:
            people.append(random.randint(5,20))
        elif hr>=9*min and hr<=12*min:
            people.append(random.randint(60,62))
        elif hr>12*min and hr<=17*min:
          people.append(random.randint(10,15))
        else:
          people.append(random.randint(0,3))
    elif weekday == 'Tuesday'or weekday == 'Thursday':
      for hr in range(hours_of_day):
         if hr>=0 and hr<7*min:
            people.append(0)
         elif hr>=7*min and hr<10*min:
            people.append(random.randint(5,10))
         elif hr>=10*min and hr<=12*min:
            people.append(random.randint(41,44))
         elif hr>12*min and hr<=15*min:
           people.append(random.randint(30,35))
         elif hr>15*min and hr<=17*min:
           people.append(random.randint(5,10))
         else:
           people.append(random.randint(0,3))
  
  #Thanksgiving break
  elif date >= datetime(2023,11,21) and date < datetime(2023,11,30):
    for hr in range(hours_of_day):
      people.append(0)  

  #Before exam week of Fall semseter
  elif date >= datetime(2023,11,30) and date < datetime(2023,12,6):
    if weekday == 'Saturday' or weekday == 'Sunday':
      for hr in range(hours_of_day):
        if hr<7*min:
          people.append(0)
        elif hr>=7*min and hr<16*min:
          people.append(random.randint(0,3))
        else:
          people.append(random.randint(0,9))    
    elif weekday == 'Monday' or weekday == 'Wednesday' or weekday == 'Friday':
      for hr in range(hours_of_day):
        if hr>=0 and hr<7*min:
            people.append(0)
        elif hr>=7*min and hr<9*min:
            people.append(random.randint(5,20))
        elif hr>=9*min and hr<13*min:
            people.append(random.randint(60,62))
        elif hr>=13*min and hr<18*min:
          people.append(random.randint(10,15))
        else:
          people.append(random.randint(0,3))
    elif weekday == 'Tuesday' or weekday == 'Thursday':
      for hr in range(hours_of_day):
         if hr>=0 and hr<7*min:
            people.append(0)
         elif hr>=7*min and hr<10*min:
            people.append(random.randint(5,10))
         elif hr>=10*min and hr<13*min:
            people.append(random.randint(41,44))
         elif hr>=13*min and hr<16*min:
           people.append(random.randint(30,35))
         elif hr>=16*min and hr<18*min:
           people.append(random.randint(5,10))
         else:
           people.append(random.randint(0,3))    

  #Fall semseter exam weeks
  elif date >= datetime(2023,12,6) and date< datetime(2023,12,17):
    if weekday == 'Saturday' or weekday == 'Sunday':
      for hr in range(hours_of_day):
        people.append(0)
    else:
      for hr in range(hours_of_day):
         if hr>=0 and hr<10*min:
            people.append(0)
         elif hr>=10*min and hr<17*min:
            people.append(random.randint(0,5))
         else:
            people.append(0)  

  #Winter break
  elif date >= datetime(2023,12,17) and date < datetime(2024,1,1): 
    for hr in range(hours_of_day):
      people.append(0) 
```

```python
start_date = datetime(2023,1,1)
end_date = start_date + timedelta(hours=8760)
time_stamp = pd.date_range(start_date, end_date, freq='T')

hrange = range(len(time_stamp))
s = time_stamp.to_series(index=hrange)
weekdays = s.dt.day_name()
row = len(hrange)
print(row)
```

```python
# add a new column "PeopleCount(p)_per_area"
df['PeoplePerArea(p/m2)'] = df['PeopleCount(p)'].apply(lambda x: x/258.65)
df['LightingPower(W)'] = df['PeopleCount(p)'].apply(lambda x: 900 if x >= 35 else (600 if x > 10 else (300 if x > 0 else 0)))
df['LightingPowerIntensity(W/m2)'] = df['LightingPower(W)'].apply(lambda x: x/258.65)
df['EquipmentLoad(W)'] = df['PeoplePerArea(p/m2)'].apply(lambda x: x/0.01*2.5865+0.3*258)
df['EquipmentPowerIntensity(W/m2)'] = df['EquipmentLoad(W)'].apply(lambda x: x/258.65)
df['VentilationRate (m3/s)'] = df['PeopleCount(p)'].apply(lambda x: x*0.01412+0.13)
df['ACH'] = df['VentilationRate (m3/s)'].apply(lambda x: x*3600/944)


df[15050:].head()
```
